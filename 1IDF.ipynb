```python
def text_process(data):
    msg=[c for c in data if c not in string.punctuation]
    msg=''.join(msg)
    msg=[word for word in msg.split() if word.lower() not in stop]
    msg=[WNL.lemmatize(word) for word in msg ]
    return msg
```
